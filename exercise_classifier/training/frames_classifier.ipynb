```python
# Load keypoints as csv
import pandas as pd


```

```python
"""
Schema of the data:
- 33 columns for the keypoints
- video_id
- frame_id within the video
- exercise name
"""
df = pd.read_csv("full_frames_dataset.csv")

df
```

```python
# Encode the labels
label_to_idx = {}

idx_to_label = {}

for label in df.exercise_label.unique():
  label_to_idx[label] = len(label_to_idx)
  idx_to_label[label_to_idx[label]] = label
```

```python
NUM_CLASSES = len(idx_to_label)

print("Number of classes", NUM_CLASSES)
```

```python
import torch
from torch import nn
```

```python
def frame_to_input(row):
  keypoints = row["left_shoulder":"right_ankle"]

  # result = 
  results = []
  for keypoint in keypoints:
    results += list(map(float, keypoints[0].replace("[", "").replace("]", "").split()))
  # 36 values for keypoints (12 * 3)
  # print(len(results))

  label = label_to_idx[row["exercise_label"]]
  return torch.Tensor(results), label


dataset = df.apply(frame_to_input, axis=1)
```

```python
print(len(dataset))
```

```python
from torch.utils.data import DataLoader , Dataset


class CustomDataset(Dataset):
  def __init__(self, dataset):
    self.dataset = dataset

  def __len__(self):
    return len(self.dataset)

  def __getitem__(self, idx):
    return self.dataset.iloc[idx]
```

```python
custom_dataset = CustomDataset(dataset)

data_loader = DataLoader(custom_dataset, batch_size=6, shuffle=True)
```

```python
import torch
from torch import nn

class SingleFrameExerciseClassifier(nn.Module):
  def __init__(self):
    super().__init__()
    self.dense = nn.Linear(36, 12)
    self.activation = nn.Tanh()
    self.dense_2 = nn.Linear(12, NUM_CLASSES)

  def forward(self, input):
    out = self.dense(input)
    out = self.activation(out)
    out = self.dense_2(out)

    return out
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
# training loop
model = SingleFrameExerciseClassifier().to(device)

optimizer = torch.optim.SGD(model.parameters(), lr=0.001)

loss = nn.CrossEntropyLoss()

def train():
  model.train()
  for batch, labels in data_loader:
    output = model(batch.to(device))

    l = loss(output, labels.to(device))

    l.backward()
    optimizer.step()
    model.zero_grad()
```

```python
def frame_accuracy(logits, labels):
  probs = nn.functional.softmax(logits, dim=1)
  preds = torch.argmax(probs, dim=1)

  # print(preds)

  return sum([1 if preds[i].item() == label.item() else 0 for i, label in enumerate(labels)])/labels.size()[0]


def evaluate():
  model.eval()
  accuracies = []
  for batch, labels in data_loader:
    output = model(batch.to(device))

    accuracies.append(frame_accuracy(output, labels.to(device)))

  mean = sum(accuracies)/len(accuracies)

  print("Mean accuracy", mean)
```

```python
evaluate()
for i in range(20):
  print("Epoch", i+1)
  train()
  evaluate()
```

```python
def predict_video_exercise(dataloader):
  model.eval()
  labels = {}
  for batch, _ in dataloader:
    logits = model(batch.to(device))

    probs = nn.functional.softmax(logits, dim=1)
    preds = torch.argmax(probs, dim=1)

    for i in range(preds.size()[0]):
      if preds[i].item() not in labels:
        labels[preds[i].item()] = 0
      labels[preds[i].item()] += 1

  video_label = max(labels, key=labels.get)

  print("Confidence : ", max(labels.values()) / sum(labels.values()))
  print(labels)
  return video_label

correct = 0
total = 249
for video_id, dl in enumerate(video_dataloaders):
  # print(len(dl))
  # print(df[df["video_id"] == video_id]["exercise_label"])
  label_name = df[(df["video_id"] == video_id) & (df["frame_id"] == 0)]["exercise_label"]
  # print(label_name.iloc[0] )
  label = label_to_idx[label_name.iloc[0] ]

  # for _, l in dl:
  #   label = l.item()
  #   break

  pred = predict_video_exercise(dl)

  print(label, pred)

  if pred == label:
    correct += 1

print("Video accuracy", correct/total)
```
